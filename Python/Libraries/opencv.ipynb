# OpenCV

```python
import cv2
import imutils
```

## Image

```python
# Read & Show data
image = cv2.imread('data/image.jpg')
cv2.imshow('Color image', image)
cv2.waitKey()

# Convert color, Show & Save img_gray
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale image', image_gray)
cv2.imwrite('data/image_gray.jpg', image)
cv2.waitKey()

# Resize data
smaller_image = cv2.resize(image, (100, 100))
cv2.imshow('Resized image', smaller_image)

# Rotate data
image_rotate = imutils.rotate(image, 90)
cv2.imshow('Rotated image', image_rotate)
cv2.waitKey()

# Threshold: Clolor data → Grayscale data → Binary data
ret, thresh_binary = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)
cv2.imshow('Binary image', thresh_binary)
cv2.waitKey()

# Canny Edge Detection: Grayscle data → Edge data
edges = cv2.Canny(image_gray, threshold1=100, threshold2=200)
cv2.imshow('Edge image', edges)
cv2.waitKey()

# Gaussian Blur
blur = cv2.GaussianBlur(image, ksize=(31, 31), sigmaX=0)    # ksize is an odd number
cv2.imshow('Blur image', blur)
cv2.waitKey()

cv2.destroyAllWindows()
```

## Camera

```python
camera_id = 0
cam = cv2.VideoCapture(camera_id)

while True:
    ret, frame = cam.read()
    if ret:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('Image from webcam', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```

## Video

```python
video_path = 'video.mp4'
video = cv2.VideoCapture(video_path)

rotate = 0
while True:
    ret, frame = video.read()
    if ret:
        if rotate != 0:
            frame = imutils.rotate(frame, rotate)
        cv2.imshow('Image from video', frame)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    elif key == ord('a'):
        rotate = 90
    elif key == ord('d'):
        rotate = -90

video.release()
cv2.destroyAllWindows()
```
